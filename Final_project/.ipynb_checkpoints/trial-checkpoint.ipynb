```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
from datetime import datetime, timedelta
```

```python
mobile = pd.read_csv('datasets/mobile_clean.csv')

mobile['event_time'] = pd.to_datetime(mobile['event_time'])
mobile['dt'] = pd.to_datetime(mobile['event_time']).dt.date
```

```python
# сортируем датасет для дальнейшей группировки
mobile = mobile.sort_values(['user_id', 'event_time']) 

# подсчет временных меток (разница между двумя временами событий - предыдущим и последующим)
diff_timestamp = mobile.groupby('user_id')['event_time'].diff()

# индексы, где будет создан новый session_id
new_session = (diff_timestamp.isnull()) | (diff_timestamp > pd.Timedelta('30Min'))

# создание уникальных session_id для каждого пользователя
mobile['session_id'] = mobile.loc[new_session, ['user_id', 'event_time']] \
    .groupby('user_id').rank(method='first').astype(int)

# заполнение пустых значений предыдущим значением
mobile['session_id'] = mobile['session_id'].fillna(method='ffill').astype(int)
```

```python
# визуализация заданного датасета
def visual_event(df, title):
    df.plot(
         kind='barh',
         figsize=(10,3), 
         grid=True,
         xlabel='Название события',
         title=title
    );
```

```python
visual_event(mobile['event_name'].value_counts(), 'Количество событий в логах')
```

```python
# функцию для отображения значений на гистограмме
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
# функция для построения простой гистограммы 
def func_bar(df, x, y, title='', xlabel='Количество', step=10):
    fig = sns.barplot(data = df, x=x, y=y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('')
    plt.xticks(np.arange(0, max(df[x])+1, step))
    
    # вызов функции для отображения значений на гистограмме
    show_values(fig, "h", space=0);
```

```python
a = mobile['event_name'].value_counts().reset_index().rename(columns = {'event_name':'total'})

# создание стиля  палитры seaborn
sns.set()
sns.set_palette('husl',8)

# построение гистограммы pаспределение заведений по категориям
func_bar(a, 'total', 'index', title='Количество событий в датасете', step=5000)
```

```python
# функция для отбражения количество событий за определенный период времени
def func_hist(df, bins=28):
    plt.figure(figsize=(8,3))
    df.hist(bins=bins)
    plt.xticks(df, rotation=70)
    plt.title('Количество событий за весь период времени');
```

```python
# визуализация количеста событий в конкретную дату времени
func_hist(mobile['event_time'])
```

```python
# создание датасета с min и max временем сессии
session = (mobile
     .pivot_table(index=['user_id','session_id'], values='event_time', aggfunc=['max','min'])
     .reset_index()
    )

# нахождение времен сессии
session['session_time'] = session['max','event_time'] - session['min','event_time']

# убираем двойные названия колонок и удаляем столбцы event_time
session.columns = [col[1] if col[1] else col[0] for col in session.columns]
session = session.drop('event_time', axis=1)

session.tail()
```
