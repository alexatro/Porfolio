## Оценка корректности проведения теста

### Соответствие данных требованиям технического задания

```python
# создание датасета с участниками и группами теста recommender_system_test
test = participants.query('ab_test == "recommender_system_test"')[['user_id', 'group']]
test.tail()
```

```python
# отбираем новых пользователей, зарегистрировавшихся с 7 по 21 декабря. Регион - EU.
users_eu = users.query('"2020-12-07" <= first_date <= "2020-12-21" & region == "EU"')['user_id']

# количество новых пользователей из Европы
users_eu.shape[0]
```

```python
# количество участников теста из Европы
test_eu = test.query('user_id in @users_eu')
test_eu['user_id'].nunique()
```

```python
# доля новых пользователей из региона EU
test_eu['user_id'].nunique() / users_eu.shape[0] * 100
```

- Всего участников теста - 6701.
- Количество новых пользователей, зарегистрировавшихся с 7 по 21 декабря, регион - EU - 42340.
- Количество участников теста из Европы, зарегистрированных по ТЗ - 6351.
- Доля участников теста относительно новых пользователей из Европы соответствует ТЗ - 15%.

```python
# формирование датасета с участниками нашего теста и их событиями
test_events = (events
               .query('user_id in @test_eu["user_id"]')
               .reset_index(drop=True)
               .drop('details', axis=1)
              )

# добавления столбца с датой регистрацией участника
test_events = test_events.merge(users[['user_id','first_date']], on='user_id', how='left')

# преобразуем колонки с датами
test_events['event_dt'] = pd.to_datetime(test_events['event_dt']).dt.date
test_events['first_date'] = pd.to_datetime(test_events['first_date']).dt.date

# считаем разницу времени между событием и датой регистрацией
test_events['delta'] = test_events['event_dt'] - test_events['first_date']

test_events.tail()
```

Так как нас интересует улучшение метрики только в течении 14 дней от регистрации, мы должны отбросить события, которые произошли после 14 дней от того момента, как был зарегистрирован пользователь, который их совершил. Нам необходимо, чтобы лайфтайм событий, которые будут влиять на наш тест, укладывался в установленный горизонт событий.

```python
# фильтруем события, которые пользователи совершили после 14 дней от регистрации
test_events = (test_events
               .query('delta < "14 days"')
               .reset_index(drop=True)
              )
test_events.tail()
```

```python
# создание списка с уникальными участниками теста и их группой
test = test_eu.query('user_id in @test_events["user_id"]')
```

```python
# функция, показывающая количество участников и их распределение по группам
def func_test(df, test):
    print(f'Количество участников: {df["user_id"].nunique()}')
    print('Распределение по группам:')
    print(test['group'].value_counts())
```

```python
# функция для создания датасета c сгруппировкой и агрегирующей функцией
def group_agg(df, columns_group, columns_agg='event_name', aggfunc='count'):
    df = (df
          .groupby(columns_group)
          .agg({columns_agg: aggfunc})
          .reset_index()
          .rename(columns = {columns_agg:'total'})
          )
    return df 
```

```python
# функцию для отображения значений на гистограмме
def show_values(axs, orient='h', space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.0f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.0f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
# функция для построения гистограммы 
def func_bar(df, x, y, hue='',
             title='', xlabel='Дата', ylabel='Количество', 
             figsize=(14,5), orient='v'):
    
    plt.figure(figsize=figsize)
    if hue == '':
        fig = sns.barplot(data = df, x=x, y=y)
    else:
        fig = sns.barplot(data = df, x=x, y=y, hue=hue)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=60)
    
    # вызов функции для отображения значений на гистограмме
    show_values(fig, orient=orient, space=0);
```

```python
# создание стиля  палитры seaborn
sns.set()
sns.set_palette('husl',8)

# вызов функции построения гистограммы
func_bar(group_agg(test_events, 'event_dt'), 'event_dt', 'total',
         title='Количество событий по дням')
```

### Воронка событий

```python
"""    
        функция для создания датасета с группировкой по событиям, количеством участников, 
        с долей пользователей, которые хоть раз совершали событие,
        с долей пользователей, проходящих на следующий шаг воронки.
        Сразу меняется порядок событий.
"""
def func_conv(df):  
    
    # создание датасета с количеством участников по событиям   
    conv = ((group_agg(df, 'event_name', 'user_id', 'nunique'))
               .reindex([0,2,1,3])
               .reset_index(drop=True)
              )
    # добавление столбца с долей пользователей, которые хоть раз совершали событие
    conv['share_id'] = conv['total'] / df['user_id'].nunique() * 100
    
    # подсчет доли пользователей, проходящих на следующий шаг воронки 
    conv['share_previous'] = conv['total'] / conv['total'].shift(1) * 100

    return conv.round(1)
```

```python
# создание датасета группы А
event_a = func_conv(test_events.query('group == "A"'))
event_a
```

```python
# создание датасета группы В
event_b = func_conv(test_events.query('group == "B"'))
event_b
```

Получилось, что просмотров корзины меньше, чем покупок. Значит есть возможность покупки со страницы товара, минуя корзину.

```python
# создание колонок с группой участников 
event_a['group'] = 'A'
event_b['group'] = 'B'

# объединение датасатетов для построения общей воронки
event_ab = pd.concat([event_a, event_b], axis=0)
```

```python
# функция для построения воронки событий
def func_funnel(x, title='Воронка событий'):
    fig = px.funnel(event_ab, x=x, y='event_name', 
                color='group', title=title)
    fig.show()
```

```python
# визуализация воронки по количеству участников
func_funnel('total', title='Воронка событий по количеству участников')
```

Ожидаемый эффект не реализовался. У тестируемой группы В нет улучшений каждой метрики не менее, чем на 10%. Наоборот, метрики снизились.

### Проверка статистической разницы долей z-критерием

Формируем гипотезу для проведения анализа, используя z-критерий Фишера на проверку равенства долей между группами:
- $Н_0$ - статистически значимых различий в доле пользователей для события между группами теста - нет.
- $Н_1$ - статистически значимые различия в доле пользователей для события между группами теста - есть.

Так как получается 3 теста, делаем поправку Бонферрони - alpha / 3

```python
# создание датасета для проведения z-теста
df_exp = (test_events
          .pivot_table(index='event_name', columns='group', values='user_id',aggfunc='nunique')
          .reset_index()
          .reindex([0,2,1,3])
          .reset_index(drop=True)
         )
df_exp
```

```python
# функция для проверки, будет ли отличие между группами статистически достоверным
def z_test(group_1, group_2, alpha=0.05/3):  
    
    for i in range(1, len(df_exp.index)):
        print(f'Событие: {df_exp.iloc[i, 0]}')
        
        successes = np.array([df_exp.loc[i, group_1], df_exp.loc[i, group_2]])
        trials = np.array([df_exp.loc[0, group_1], df_exp.loc[0, group_2]])
        
        stat, p_value = proportions_ztest(successes, trials)
        
        print(successes, trials)

        print('p-значение: ', p_value)

        if p_value < alpha:
            print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
        else:
            print(
                'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
            )
        print('-'*80)
```

```python
# проведение теста
z_test('A', 'B')
```

- Статистический тест показал, что на этапе `product_page` между долями группы А и группы В - есть значимая разница.
- На остальных этапах `product_cart` и `purchase` - нет оснований считать доли групп разными.

Но в датасете нет данных о событиях, которые были после 30 декабря 2020 года. Дата остановки теста - 4 января 2021 года. Нет информации о событиях за 4 дня. Это могло повлиять на исход теста. Если учесть эти данные, то конверсия могла быть иной.
